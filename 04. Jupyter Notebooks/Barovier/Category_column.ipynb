```python
df1 = pd.concat([df, df.categories.str.get_dummies(sep=",")], axis=1)
```

```python
#Estimating the time to shutdown (0 output) after the first Yelp review 
```

```python
#Objectives
#(1) What is the impact of Yelp star reviews on restaurants' survival? Does star ratings predict survival?
#For example, is there any difference between restaurants with average star reviews of 1 nd 5?
#(2) What is the probability that a restaurant survives a specific time (years)? 
#For example, given an average star review, we will tell that if 5 years after
#has been passed, then the probability of that restaurant avoiding shut down at that time will be ~0.7, (and also given star rating).
#(3)Are there differences in survival between groups of restaurants? 
#For example, Let’s say there are two groups of restaurants: chain and independent. 
#Those two groups received different star reviews. 
#Our goal here will be to find out if there is a significant difference between the survival time for 
#those two different groups based on the star reviews they were given.
```

```python
#Categorizing chain and independent restaurants; if name > 2 == chain 
df2 = df1['name'].value_counts() > 2
df2 = pd.DataFrame(list(df2))
```

```python
df1 = pd.concat([df1, df2.reindex(df1.index)], axis=1)
```

```python
df1 = df1.rename(columns={0: 'is_chain'})
df1 = df1.dropna(subset=['is_chain']) #removes 160 rows, only for survival data analysis
df1["is_chain"] = df1["is_chain"].astype(int)
df1.head()
```

```python
#Import required libraries:
#pip install lifelines
#conda install -c conda-forge lifelines
import numpy as np
import matplotlib.pyplot as plt
from lifelines import KaplanMeierFitter
```

```python
#Statistical info about our dataset:
columns = ['Unnamed: 0.1']
df1.drop(columns, inplace=True, axis=1)
df1.describe()
```

```python
#Plot histogram for is_chain distribution:
print (df1["is_chain"].hist())
```

```python
#Plot histogram for star review distribution:
print (df1["stars"].hist())
```

```python
#Create an object of KaplanMeierFitter:
kmf = KaplanMeierFitter()
```

```python
#Fit the parameter values in our object:
kmf.fit(durations =  df1["yelp_lifetime"], event_observed = df1["is_open"])
```

```python
#Print the event table:
kmf.event_table.head()
# Removed = Observed + Censored
# Censored = Restaurant that didn't close.(They are of no use to us!)
# Observed = Restaurant that closed.
```

```python
#event_at = It stores the value of survival days for the subjects.
#at_risk = number of restaurants under observation
#entrance = restaurants added after a certain timeframe 
#censored = restaurant that survives at the end of our experiment 
#observed = number of restaurants that close at the end of experiment 
#removed = restaurants that are no longer apart of our experiment 
```

```python
#Calculating the survival probability for a given time:
event_at_0 = kmf.event_table.iloc[0,:]

#Calculate the survival probability for t=0:
surv_for_0 = (event_at_0.at_risk - event_at_0.observed)/event_at_0.at_risk
surv_for_0

#note this is not the probability for the entire time for a restaurant
#In a nutshell, we want to find the probability of a person surviving all of the time he lived after diagnosis
#What we just found is the probability of a particular experiment only.
#--For instance, we know that a restaurant has survived the 1st time interval, 
#and we want to find the probability of it surviving the second time interval given that it
#has survived the 1st time interval. (conditional probability)
```

```python
#Calculating the survival probability for a given time:
event_at_13 = kmf.event_table.iloc[1,:]

#Calculate the survival probability for t=0:
surv_for_13 = (event_at_13.at_risk - event_at_13.observed)/event_at_13.at_risk
surv_for_13
```

```python
#Calculating the actual survival probability at a given time:

surv_after_0 = surv_for_0 
print("Survival Probability After 0 Days: ",surv_after_0)
```

```python
#Calculating the actual survival probability at a given time:

surv_after_13 = surv_for_13
print("Survival Probability After 0 Days: ",surv_after_13)
```

```python
#Calculating the actual survival probability at a given time:
surv_after_13 = surv_for_0 * surv_for_13
print("Survival Probability After 13 Days: ",surv_after_13)
```

```python
#Get the probability values the easy way!
print("Survival probability for t=0.000000: ",kmf.predict(0.000000))
print("Survival probability for t=0.041096: ",kmf.predict(0.041096))
print("Survival probability for t=6: ",kmf.predict(3)) #probability of surviving for 3 years
```

```python
#Predicting the surviaval probability for an array of value:
kmf.predict([0,5,11,12])
```

```python
#To get the full list:
kmf.survival_function_.head()
```

```python
#Plot the graph:
kmf.plot()
plt.title("The Kaplan-Meier Estimate")
plt.xlabel("Number of years")
plt.ylabel("Probability of survival")
```

```python
#Survival probability with confidence interval:
kmf.confidence_interval_survival_function_.head()
```

```python
#Plot survival function with confidence interval:
confidence_surv_func = kmf.confidence_interval_survival_function_
plt.plot(confidence_surv_func["KM_estimate_lower_0.95"],label="Lower")
plt.plot(confidence_surv_func["KM_estimate_upper_0.95"],label="Upper")
plt.title("Survival Function With Confidence Interval")
plt.xlabel("Number of years")
plt.ylabel("Survival Probability")
plt.legend()
```
